```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv("Coffee_Data.csv")
```

```python
data = df["age"].value_counts()

plt.figure(figsize=(8,5))
data.plot(kind="bar")
plt.title("Coffee drinkers by age group")
plt.xlabel("Age Group")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
data = df["gender"].value_counts()

plt.figure(figsize=(8,5))
data.plot(kind="bar")
plt.title("Coffee drinkers by gender group")
plt.xlabel("gender Group")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
data = df["education_level"].value_counts()
plt.figure(figsize=(8,5))
plt.title("coffee drinkers by education level")
data.plot(kind="bar")
plt.xlabel("Education Level")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
data = df["employment_status"].value_counts()
plt.figure(figsize=(8,5))
plt.title("coffee drinkers by employment status")
data.plot(kind="bar")
plt.xlabel("employment status ")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
coffee_types = (
    df["brew"]
    .str.split(",")
    .explode()
    .str.strip()
)
coffee_counts = coffee_types.value_counts()

coffee_counts.head(10)

plt.figure(figsize=(10,5))
coffee_counts.plot(kind="bar")
plt.title("Most Preferred Coffee Types")
plt.xlabel("Coffee Type")
plt.ylabel("Count")
plt.xticks(rotation=90)
for i, v in enumerate(coffee_counts):
    plt.text(i,v,str(v),ha='center')
plt.show()
```

```python
data=df["roast_level"].value_counts()
plt.figure(figsize=(8,5))
data.plot(kind="bar")
plt.xlabel("Roast Level")
plt.ylabel("Count")
plt.xticks(rotation=45)
for i, v in enumerate(data):
    plt.text(i,v,str(v),ha='center')
plt.title("Most preferred roast level type")
plt.show()
```

```python
ax=pd.crosstab(df["age"], df["caffeine"]).plot(
    kind="bar",
    figsize=(10,6)
)

plt.xlabel("Age Group")
plt.ylabel("Count")
plt.title("Preferred Caffeine Level by Age")
plt.xticks(rotation=45)
for container in ax. containers:
    ax.bar_label(container,padding=3)
plt.show()
```

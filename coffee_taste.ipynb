```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv("Coffee_Taste_Test.csv")
print(f"Rows * columns are : {df.shape}")#rows * columns (4042 * 57)
df.columns 
df.head(5)
df.tail(5)
df.dtypes
df.info()
df.describe()
df.duplicated().sum()
df[["min_age","max_age"]] = (
    df["age"]
    .str.replace(">","")
    .str.replace("<","")
    .str.replace("years old", "")
    .str.split("-", expand=True)
)
df["cups"] = df["cups"].str.replace("More than","")
df["min_age"] = pd.to_numeric(df["min_age"])
df["max_age"] = pd.to_numeric(df["max_age"])
df["avg_age"] = df[["min_age","max_age"]].mean(axis=1)
df.drop("max_age",axis=1,inplace=True)
df.drop("min_age",axis=1,inplace=True)
df["cups"]=df["cups"].bfill().str.replace("Less than","")
df["where_drink"] = df["where_drink"].bfill(inplace=True)
df["brew"]=df["brew"].bfill(inplace=True)
df["brew_other"]=df["brew_other"].bfill(inplace=True)
df["purchase"]=df["purchase"].bfill(inplace=True).ffill(inplace=True)
df["purchase_other"]=df["purchase_other"].bfill(inplace=True).ffill(inplace=True)
df["favorite"]=df["favorite"].ffill(inplace=True)
df["favorite_specify"]=df["favorite_specify"].bfill(inplace=True).ffill(inplace=True)
df["value_equipment"]=df["value_equipment"].bfill(inplace=True)
df["gender"]=df["gender"].bfill(inplace=True)
df["gender_specify"]=df["gender_specify"].bfill(inplace=True).ffill(inplace=True)
df["education_level"]=df["education_level"].bfill(inplace=True)
df["ethnicity_race"]=df["ethnicity_race"].bfill(inplace=True)
df["ethnicity_race_specify"]=df["ethnicity_race_specify"].bfill(inplace=True).ffill(inplace=True)
df["employment_status"]=df["employment_status"].bfill(inplace=True)
df["number_children"]=df["number_children"].bfill(inplace=True)
df["political_affiliation"]=df["political_affiliation"].bfill(inplace=True)
df["age"]=df["age"].ffill(inplace=True)
df["additions"]=df["additions"].ffill(inplace=True)
df["additions_other"]=df["additions_other"].bfill(inplace=True).ffill(inplace=True)
df["dairy"]=df["dairy"].bfill(inplace=True)
df["sweetener"]=df["sweetener"].bfill(inplace=True)
df["style"]=df["style"].bfill(inplace=True)
df["strength"]=df["strength"].bfill(inplace=True)
df["roast_level"]=df["roast_level"].bfill(inplace=True)
df["caffeine"]=df["caffeine"].bfill(inplace=True)
df["expertise"]=df["expertise"].bfill(inplace=True)
df["coffee_a_bitterness"]=df["coffee_a_bitterness"].bfill(inplace=True)
df["coffee_a_acidity"]=df["coffee_a_acidity"].bfill(inplace=True)
df["coffee_a_personal_preference"]=df["coffee_a_personal_preference"].bfill(inplace=True)
df["coffee_a_notes"]=df["coffee_a_notes"].ffill(inplace=True).bfill(inplace=True)
df["coffee_b_bitterness"]=df["coffee_b_bitterness"].bfill(inplace=True)
df["coffee_b_acidity"]=df["coffee_b_acidity"].bfill(inplace=True)
df["coffee_b_personal_preference"]=df["coffee_b_personal_preference"].bfill(inplace=True)
df["coffee_b_notes"]=df["coffee_b_notes"].bfill(inplace=True)
df["coffee_c_bitterness"]=df["coffee_c_bitterness"].bfill(inplace=True)
df["coffee_c_acidity"]=df["coffee_c_acidity"].bfill(inplace=True)
df["coffee_c_notes"]=df["coffee_c_notes"].bfill(inplace=True)
df["coffee_c_personal_preference"]=df["coffee_c_personal_preference"].bfill(inplace=True)
df["coffee_d_acidity"]=df["coffee_d_acidity"].bfill(inplace=True)
df["coffee_d_bitterness"]=df["coffee_d_bitterness"].bfill(inplace=True)
df["coffee_d_personal_preference"]=df["coffee_d_personal_preference"].bfill(inplace=True)
df["coffee_d_notes"]=df["coffee_d_notes"].bfill(inplace=True)
df["prefer_abc"]=df["prefer_abc"].bfill(inplace=True)
df["prefer_ad"]=df["prefer_ad"].bfill(inplace=True)
df["prefer_overall"]=df["prefer_overall"].bfill(inplace=True)
df["wfh"]=df["wfh"].bfill(inplace=True)
df["total_spend"]=df["total_spend"].bfill(inplace=True)
df["wfh"]=df["wfh"].bfill(inplace=True)
df["why_drink"]=df["why_drink"].bfill(inplace=True)
df["why_drink_other"]=df["why_drink_other"].bfill(inplace=True).ffill(inplace=True)
df["taste"]=df["taste"].bfill(inplace=True)
df["know_source"]=df["know_source"].bfill(inplace=True)
df["most_paid"]=df["most_paid"].bfill(inplace=True)
df["most_willing"]=df["most_willing"].bfill(inplace=True)
df["value_cafe"]=df["value_cafe"].bfill(inplace=True)
df["spent_equipment"]=df["spent_equipment"].bfill(inplace=True)
df["value_cafe"]=df["value_cafe"].bfill(inplace=True)
df["avg_age"] = df["avg_age"].ffill()
df[["less_spend","High_spend"]]=df["total_spend"].str.replace("$","").str.replace(">","").str.replace("<","").str.split("-",expand=True)
df["less_spend"] = pd.to_numeric(df["less_spend"])
df["High_spend"] = pd.to_numeric(df["High_spend"])
df["spend"]=df[["less_spend","High_spend"]].mean(axis=1)
df.drop("High_spend",axis=1,inplace=True)
df.drop("less_spend",axis=1,inplace=True)
df[["less_spend_equipment","High_spend_equipment"]]= df["spent_equipment"].str.replace("$","").str.replace(">","").str.replace("<","").str.replace("More than","").str.replace(",","").str.replace("Less than","").str.split("-",expand=True)
df["less_spend_equipment"] = pd.to_numeric(df["less_spend_equipment"])
df["High_spend_equipment"] = pd.to_numeric(df["High_spend_equipment"])
df["spend_on_equipment"]=df[["less_spend_equipment","High_spend_equipment"]].mean(axis=1)
df.drop("High_spend_equipment",axis=1,inplace=True)
df.drop("less_spend_equipment",axis=1,inplace=True)
df[["willing1","willing2"]]= df["most_willing"].str.replace("$","").str.replace(">","").str.replace("<","").str.replace("More than","").str.replace(",","").str.replace("Less than","").str.split("-",expand=True)
df["willing1"] = pd.to_numeric(df["willing1"])
df["willing2"] = pd.to_numeric(df["willing2"])
df["most_willing"]=df[["willing1","willing2"]].mean(axis=1)
df.drop("willing2",axis=1,inplace=True)
df.drop("willing1",axis=1,inplace=True)
df.isna().sum()
df.to_csv("Coffee_Data.csv", index=False)
print("file cleaned successfully!")
```
